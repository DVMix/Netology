9.Сделать кросс-валидациюданных с использованием подхода K - fold (n_folds=10). 

```python
# http://www.alfredo.motta.name/cross-validation-done-wrong/
from pandas import *
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import *
from sklearn.metrics import *
import random
```

```python
kf = StratifiedKFold(data_r['diagnosis_is_B'], n_folds=10)

mse = []
fold_count = 0
for train, test in kf:
    train_fold = data_r.ix[train]
    test_fold = data_r.ix[test]
 
    # find best features
    corr = train_fold.corr()['diagnosis_is_B'][train_fold.corr()['diagnosis_is_B'] < 1].abs()
    corr = corr.sort_values(ascending=False)
    features = corr.index[:9].values
  
    # Get training examples
    train_fold_input = train_fold[features].values
    train_fold_output = train_fold['diagnosis_is_B']
 
    # Fit logistic regression
    logreg = LogisticRegression()
    logreg.fit(train_fold_input, train_fold_output)
 
    # Check MSE on test set
    pred = logreg.predict(test_fold[features])
    mse.append(mean_squared_error(test_fold.diagnosis_is_B, pred))
     
    # Done with the fold
    fold_count += 1
    
print(DataFrame(mse).mean())
```

11.Решить задачу бинарной классификации и предсказать переменную ”Diagnosis” протестировав как минимум 2 
алгоритма. Использовать те алгоритмы,которые позволяют предсказать вероятность класса (proba). 
Рассчитать и вывести вероятность каждого класса. 

```python
classificaation_data = data_r[col_list]
classificaation_data.head()
classificaation_target = data_r['diagnosis_is_B']

from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(classificaation_data, classificaation_target, 
                                                    test_size=0.2, random_state=42)

print(train_X.shape)
print(test_X.shape)
print(train_y.shape)
print(test_y.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, max_depth=5,random_state=0)
classifier.fit(train_X, train_y)

predicted_y = classifier.predict(test_X)
pred_proba_y = classifier.predict_proba(test_X)

classifier.score(test_X, test_y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, 
                                 max_depth=1, random_state=0).fit(train_X, train_y) 
predicted_y_GBC = clf.predict(test_X)
pred_proba_y_GBC = clf.predict_proba(test_X)

clf.score(test_X, test_y) 
```

12.Проверить качество классификации с использованием следующих метрик: Accuracy, F1-Score, Precision, Recall

```python
# accuracy - https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html

import numpy as np
from sklearn.metrics import accuracy_score

y_pred = predicted_y
y_true = test_y
print('    accuracy_score   = ',accuracy_score(y_true, y_pred))
print('    accuracy_score_N = ',accuracy_score(y_true, y_pred, normalize=False))

y_pred = predicted_y_GBC
print('GBS_accuracy_score   = ',accuracy_score(y_true, y_pred))
print('GBC_accuracy_score_N = ',accuracy_score(y_true, y_pred, normalize=False))
```

```python
# F1-Score - https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html

from sklearn.metrics import f1_score
y_true = test_y
y_pred = predicted_y

print('f1_Score_macro    = ', f1_score(y_true, y_pred, average='macro') )
print('f1_Score_micro    = ', f1_score(y_true, y_pred, average='micro') )
print('f1_Score_weighted = ', f1_score(y_true, y_pred, average='weighted') )
print('f1_Score_none     = ', f1_score(y_true, y_pred, average=None) )
print('/n')
y_pred = predicted_y_GBC
print('GBC_f1_Score_macro    = ', f1_score(y_true, y_pred, average='macro') )
print('GBC_f1_Score_micro    = ', f1_score(y_true, y_pred, average='micro') )
print('GBC_f1_Score_weighted = ', f1_score(y_true, y_pred, average='weighted') )
print('GBC_f1_Score_none     = ', f1_score(y_true, y_pred, average=None) )
```

```python
# Precision- https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_score.html
from sklearn.metrics import precision_score
y_true = test_y
y_pred = predicted_y

print('precision_macro       = ', precision_score(y_true, y_pred, average='macro')  )
print('precision_micro       = ', precision_score(y_true, y_pred, average='micro')  )
print('precision_weighted    = ', precision_score(y_true, y_pred, average='weighted') )
print('precision_none        = ', precision_score(y_true, y_pred, average=None)   )

y_pred = predicted_y_GBC
print('GBC_precision_macro       = ', precision_score(y_true, y_pred, average='macro')  )
print('GBC_precision_micro       = ', precision_score(y_true, y_pred, average='micro')  )
print('GBC_precision_weighted    = ', precision_score(y_true, y_pred, average='weighted') )
print('GBC_precision_none        = ', precision_score(y_true, y_pred, average=None)   )
```

```python
# Recall   -https://scikit-learn.org/stable/modules/generated/sklearn.metrics.recall_score.html
from sklearn.metrics import recall_score
y_true = test_y
y_pred = predicted_y

print('recall_macro    = ', recall_score(y_true, y_pred, average='macro'))  
print('recall_micro    = ', recall_score(y_true, y_pred, average='micro'))  
print('recall_weighted = ', recall_score(y_true, y_pred, average='weighted'))  
print('recall_none     = ', recall_score(y_true, y_pred, average=None))  

y_pred = predicted_y_GBC
print('GBC_recall_macro    = ', recall_score(y_true, y_pred, average='macro'))  
print('GBC_recall_micro    = ', recall_score(y_true, y_pred, average='micro'))  
print('GBC_recall_weighted = ', recall_score(y_true, y_pred, average='weighted'))  
print('GBC_recall_none     = ', recall_score(y_true, y_pred, average=None))
```
