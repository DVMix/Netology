```python
# 6 - Попробовать стратегию oversamling и сравнить ее с неизмененной моделью
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from imblearn.over_sampling import SMOTE
```

```python
training_features, test_features, training_target, test_target, = train_test_split(X,y,test_size = .1, random_state=12)
```

```python
x_train, x_val, y_train, y_val = train_test_split(training_features, training_target,test_size = .1,random_state=12)
```

```python
sm = SMOTE(random_state=12, ratio = 1.0)
x_train_res, y_train_res = sm.fit_sample(x_train, y_train)
```

```python
clf_rf = RandomForestClassifier(n_estimators=25, random_state=12)
clf_rf.fit(x_train_res, y_train_res)
```

```python
clf_rf.score(x_train_res, y_train_res)
```

```python
classifier_fs1 = RandomForestClassifier()
classifier_fs1.fit(x_train_res, y_train_res)
classifier_fs1.score(x_train_res, y_train_res)
```

```python
clf_rf.score(x_train_res, y_train_res) - classifier_fs1.score(x_train_res, y_train_res)
```

После OVERSAMPLING'а получили разницу вв 3-м знаке

```python
recall_score(y_val, clf_rf.predict(x_val))
```

```python
recall_score(y_val, classifier_fs1.predict(x_val))
```
