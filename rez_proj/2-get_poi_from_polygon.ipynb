```python
import os
import osmnx as ox
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from osmnx import pois
```

```python
city = 'Nizhny Novgorod'
```

```python
mos_poly = ox.gdf_from_place(city+', Russian Federation',which_result=2)
mos_poly
```

```python
mos_poly.geometry[0]
```

```python
mos_pois = pois.pois_from_polygon(mos_poly.geometry[0])
```

```python
columns = ['osmid','name','amenity','geometry']
rez_gdf = mos_pois[columns]
```

```python
print(rez_gdf.shape)
rez_gdf.head()
```
