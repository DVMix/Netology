# 2. METRO

```python
def getkeys(source):
    _keylist = []
    source = source.tolist() 
    for element in source:
        for key in element.keys():
            if key not in _keylist:
                _keylist.append(key)
    return _keylist
```

```python
new_df = pd.DataFrame()
col_name = 'metro'
proc_list = []
proc_list=getkeys(df[col_name])
proc_list.sort()

for new_column in proc_list:
    #print(new_column)
    new_column_list = []
    for line in df[col_name]:
        #print(line)
        if new_column in line.keys():
            if line[new_column]!= 'nan':
                value = int(re.compile("[0-9]+").findall(line[new_column])[0])
            else:
                value = 0
            new_column_list.append(value)
        else:
            new_column_list.append(0)
    tmp = pd.DataFrame({new_column : new_column_list})
    new_df= pd.concat([new_df,tmp],axis=1) 
metro_df = new_df
del new_df
```

```python
def plosyad(data):
    rez = re.compile("[0-9,0-9]+").findall(data.replace(' ','').replace(',','.'))[-1]
    return float(rez)

def etazh(data):
    if type(data) == float:
        rez = 1
    else:
        rez = int(re.compile("[0-9]+").findall(data)[0])
    return rez

def etazhnost(data):
    if type(data) != float:
        #rez = int(re.compile("[0-9]+").findall(data)[1])
        rez = int(re.compile("[0-9]+").findall(data)[-1])
    else:
        rez = 1    
    return rez

def price(data):
    rez = re.compile("[0-9]+").findall(data.replace(' ',''))[0]
    return int(rez)

def stavka(data):
    rez = re.compile("[0-9]+").findall(data.replace(' ',''))[0]
    return int(rez) 

def tax_type(data):
    try:
        r = data.split(':')[0]
    except:
        r = 'No info'
    return r

def tax_value(data):
    try:
        v = int(data.split(':')[1].replace(' ','').replace('₽',''))
    except:
        v = 0
    return v

def o_platezh(data):
    if type(data)!=float:
        if data !='-':
            return  int(data.replace(' ','').replace('₽',''))
        else:
            return 0
    else:
        return 0
def parking(data):
    if type(data) == float:
        rez = 0
    else:
        if (data == '-') | (data =='Бесплатно'):
            rez = 0
        else:
            rez = int(data.replace(' ','').replace('₽/место','').replace('%',''))
    return rez

def nan_zero(data):
    if type(data) != float:
        return int(data)
    else:
        return 0

def square_place(data):
    if type(data)!= float:
        rez = float('.'.join(re.compile("[0-9,0-9]+").findall(data.replace(' ','').replace(',','.'))))
    else:
        rez = 0
    return rez
def potolok(data):
    if type(data)!= float:
        rez = float('.'.join(re.compile("[0-9,0-9]+").findall(data.replace(' ','').replace(',','.'))))
    else:
        rez = 2.75
    return rez
def infrastructure(data): 
    if type(data) == float:
        rez = 'Нет информации'
    else:
        for el in data:
            if el not in rez_list:
                rez_list.append(el)
```
