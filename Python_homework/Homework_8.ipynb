```python
# Загрузить файл wines_data.csv
import pandas as pd
```

```python
wines = pd.read_csv('wine_data.csv')
wines.head()
```

```python
# Рассчитать и визуализировать корреляции для переменных данного датасета
```

```python
from scipy import stats
import seaborn as sns
```

```python
corr = wines.corr()
corr
```

```python
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
# Визуализировать взаимосвязи между 3-4 переменными
wines.head()
```

```python
sns.pairplot(wines, vars=['Alcohol', 'Alcalinity of ash'],
                 kind='scatter')  
```

```python
sns.pairplot(wines, vars=['Nonflavanoid phenols', 'Magnesium'],
                 kind='scatter')  
```

```python
sns.pairplot(wines, vars=['Magnesium', 'Proline'],
                 kind='scatter')  
```

```python
# Рассчитать t-test статистики для 3-4 переменных
```

```python
stats.ttest_rel(wines['Alcohol'], wines['Alcalinity of ash'])  
```

```python
stats.ttest_rel(wines['Nonflavanoid phenols'], wines['Magnesium'])   
```

```python
stats.ttest_rel(wines['Magnesium'], wines['Proline'])   
```
