```python
# Задание 2
# Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
# - оценка 2 и меньше - низкий рейтинг
# - оценка 4 и меньше - средний рейтинг
# - оценка 4.5 и 5 - высокий рейтинг

# Результат классификации запишите в столбец class
```

```python
# функция для поиска соответствий
def class_x(x):
    
    # переменная строкового типа для записи в неё значения классификатора
    rez = ''
    if(x<=2):
        rez = 'low rating'
    else:
        if(x<=4):
            rez = 'mid rating'
        else:
            rez = 'high rating'
    return rez

# функция добавляет столбец с классификацией по рейтингу
def classificator(data):
    data['class'] = data['rating'].apply(lambda x: class_x(x))
    return data   

# считываю данные из файла в датафрейм
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

# создаю сводную таблицу
joined = movies.merge(ratings, on='movieId')

# использую функцию классификатора
a = classificator(joined)
a.head()
```

```python
# Задание 3
# Посчитайте среднее значение Lifetime киноманов (пользователи, которые поставили 100 и более рейтингов). 
# Под Lifetime понимается разница между максимальным и минимальным значением timestamp для каждого 
# пользователя. Ответ дайте в днях.
```

```python
# считаею количество оценок каждого пользователя
num_of_rates = ratings.groupby('userId').agg({'timestamp':'count'}).reset_index()

# записываю пользователей, у которых болььше 100 оценок
more_than_100_rates = num_of_rates[num_of_rates.timestamp>=100]

# назначаю удобные названия колнкам
more_than_100_rates.columns = ['userId','num_of_rates']

# дополняю таблицу значениями из ratings
more_than_100_rates_Full = more_than_100_rates.merge(ratings, on='userId', how='left')

# ищу для каждого пользователя последнее время оценки на сайте
fresh_timestamp = more_than_100_rates_Full.groupby('userId').agg({'timestamp':['min']}).reset_index()

# назначаю удобные названия колнкам
fresh_timestamp.columns = ['userId','l_time']

#дополняю полную таблицу колонкой времен, когда пользователь последниц раз ставил оценки
final_table = more_than_100_rates_Full.merge(fresh_timestamp, on='userId', how='left')

# вычитаю значение времени последнего оценивания из времени самого давнего оценивания 
final_table['difference'] = final_table['timestamp'] - final_table['l_time']

# ищу среднее значение полученных результатов для каждого пользователя
final_table_average_mean = final_table.groupby('userId').agg({'difference':'mean'}).reset_index()

# создаю еще 1 колонку для удобства, в которую запишу эти данные в виде полных дней
final_table_average_mean['in_days'] = final_table_average_mean['difference'].apply(lambda x: int(x/86400))

# вывод на экран
final_table_average_mean.head()
```
